# Introduction

## System

The fourth project that you will complete this semester is to design, implement, and test a controller that enables a quadrotor to race through gates from start to finish without crashing:

In particular, your drone will be racing with other drones. You will need to take care not to run into these other drones, as collisions may cause drones to crash.

## Context

Imagine that, working as a control systems engineer, you have been hired by [Tiny Whoop](https://www.tinywhoop.com/) to design a controller for a small-scale drone. In particular, imagine that Tiny Whoop intends to market this drone to amateur pilots for racing (e.g., see their [alley cat coffee cup invitational race](https://www.youtube.com/watch?v=jnXgoRxyVqU&t=283s)). Your job is to show that this drone  with a suitable controller  is capable of high-speed, agile flight.

As you work on this project, we encourage you to think about other possible applications of drones such as these[. PowderBee from Bluebird Mounta](https://snowbrains.com/new-avalanche-rescue-drone-in-the-works/)in, for example, is intended t[o find avalanche victi](https://www.springwise.com/innovation/sport-fitness/powderbee-drone-avalanche-rescue/)ms (i.e., people buried under thick snow) quickly so they can be rescued before succumbing to asphyxiation, hypothermia, or other injuries  every second counts in rescue applications, just as for racing drones.
